```python
# Select data for states
from datetime import datetime
df = pd.read_csv("2020_MX_Region_Mobility_Report.csv", encoding='cp1252')
df = df.rename(columns={'retail_and_recreation_percent_change_from_baseline': 'Retail', 'grocery_and_pharmacy_percent_change_from_baseline': 'Grocery', 'parks_percent_change_from_baseline':'Parks', 'transit_stations_percent_change_from_baseline':'Transit', 'workplaces_percent_change_from_baseline':'Workplaces', 'residential_percent_change_from_baseline':'Residential'})
edos = df[df.sub_region_1.notnull()]
edos
```

```python
# Extract state names
edo_names = edos['sub_region_1'].unique()
edo_names
```

```python
# Select only residential data
edos = edos[['date','sub_region_1','Residential']]
edos['date']=pd.to_datetime(edos['date'])
edos
```

```python
## Make a time series for each state
series = []
for edo in edo_names:
    serie = edos[edos['sub_region_1']==edo]
    serie = serie[['date','Residential']]
    serie.set_index('date',inplace=True)
    ts = serie['Residential']
    series.append(ts)
```
